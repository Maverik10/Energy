```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Dense
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
df = pd.read_csv('dataset - dataset.csv.csv')
```

```python
df['date'] = pd.to_datetime(df['date'], format="%d-%m-%Y %H:%M", errors='coerce')
```

```python
df['T1_rolling_mean'] = df['T1'].rolling(window=12).mean()
df['RH_1_rolling_mean'] = df['RH_1'].rolling(window=12).mean()
```

```python
df.fillna(method='bfill', inplace=True) 
```

```python
df1 = df[['Appliances', 'T1', 'RH_1', 'Press_mm_hg', 'RH_out', 'Windspeed', 'T1_rolling_mean', 'RH_1_rolling_mean']]
```

```python
scaler = StandardScaler()
df1_scaled = scaler.fit_transform(df1)
```

```python
X = df1_scaled
y = df1_scaled[:, 0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, shuffle=False)
```

```python
Time_stamps = 100 
features = X.shape[1]
```

```python
Train_set = TimeseriesGenerator(X_train, y_train, length=Time_stamps, batch_size=32)
Test_set = TimeseriesGenerator(X_test, y_test, length=Time_stamps, batch_size=32)
```

```python
model = Sequential([
    LSTM(150, input_shape=(Time_stamps, features), return_sequences=True),
    Dropout(0.3),
    LSTM(150, return_sequences=True),
    Dropout(0.3),
    LSTM(100),
    Dropout(0.3),
    Dense(1)
])
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
```

```python
history = model.fit(Train_set, epochs=200, validation_data=Test_set, shuffle=False)
```

```python
predictions = model.predict(Test_set)

# Rescale predictions and y_test to original scale
y_test_rescaled = scaler.inverse_transform(np.concatenate((y_test.reshape(-1, 1), X_test[:, 1:]), axis=1))[:, 0]
predictions_rescaled = scaler.inverse_transform(np.concatenate((predictions, X_test[Time_stamps:, 1:]), axis=1))[:, 0]
```

```python
y_test_rescaled = y_test_rescaled[-len(predictions_rescaled):]

# Performance Metrics
mse = mean_squared_error(y_test_rescaled, predictions_rescaled)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test_rescaled, predictions_rescaled)
mape = np.mean(np.abs((y_test_rescaled - predictions_rescaled) / y_test_rescaled)) * 100
accuracy = 100 - mape

# Print results
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("Mean Absolute Error:", mae)
print("Mean Absolute Percentage Error:", mape)
print("Accuracy:", accuracy, "%")

# Plotting Actual vs Predicted values
plt.figure(figsize=(12, 8))
plt.plot(y_test_rescaled, label="Actual Value")
plt.plot(predictions_rescaled, label="Predicted Value")
plt.legend()
plt.show()
```
